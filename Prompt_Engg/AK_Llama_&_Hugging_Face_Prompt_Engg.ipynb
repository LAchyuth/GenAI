# Loading Llama 2-7B

```python
from transformers import AutoTokenizer  #consolidate package to use Bert,GPT,roberta etc
import transformers
import torch #instead of keras and pytorch as this is Llama model

# Specify the model to be used
model = "meta-llama/Llama-2-7b-chat-hf"
```

```python
# Load the pretrained tokenizer
# tokenizer = AutoTokenizer.from_pretrained(model)

tokenizer = AutoTokenizer.from_pretrained(model,use_auth_token = True)
```

```python
# Make it ready for text generation

pipeline = transformers.pipeline(
          "text-generation",
          model = model,
          torch_dtype = torch.float16,
          device_map = "auto"
)
```

```python
def get_response(prompt):

  sequences = pipeline(prompt,
      do_sample=True,
      return_full_text=False,
      top_k=10,
      #top_p = 0.9,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
      max_length=1000,
  )

  return sequences[0]['generated_text']
```

```python
def get_response(prompt):
  sequences = pipeline(prompt,
                  do_sample =True, # by default it will consider nearest word
                  top_k = 10,  # for predicting select top 10 values
                  #top_p = 0.9 # for predicting use words which are cummulative of 90%
                  num_return_sequences = 1,  # no of outputs
                  eos_token_id = tokenizer.eos_token_id, #
                  max_length = 1000   # max tokens generated
  )

  return sequences[0]['generated_text']
```

# Prompting Llama - 2

## Prompt Template


### template to follow to write a prompt in Llama
<s>[INST] <<SYS>>
{{ system_prompt }}
<</SYS>>

{{ user_message }} [/INST] {{model_answer}} </s>


```python
prompt1 = '''

<s>[INST] <<SYS>>
You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe.
Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content.
Please ensure that your responses are socially unbiased and positive in nature.
If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct.
If you don't know the answer to a question, please don't share false information.
<</SYS>>
I liked friends and money heist. Recommend me the similar shows to watch.
[/INST]
'''
```

```python
print(get_response(prompt1))
```
